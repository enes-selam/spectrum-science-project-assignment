# Tayf Bilimine Giriş Proje Ödevi 3: Çift Yıldızlarda Tayfsal Analiz

```python
import numpy as np
import openpyxl
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data = openpyxl.load_workbook("proje_ödev_3_veri.xlsx")
spec_data = data["Tayf_Veri"]
```

```python
lam, spec_1, spec_2, spec_3, spec_4, spec_5, spec_6 = [], [], [], [], [], [], [] 
i = 4
while i < 5004:
    lam.append(spec_data.cell(i,1).value)
    spec_1.append(spec_data.cell(i,2).value)
    spec_2.append(spec_data.cell(i,3).value)
    spec_3.append(spec_data.cell(i,4).value)
    spec_4.append(spec_data.cell(i,5).value)
    spec_5.append(spec_data.cell(i,6).value)
    spec_6.append(spec_data.cell(i,7).value)
    i += 1
```

```python
lam_1, spec_1_1, spec_2, spec_3, spec_4, spec_5, spec_6 = np.array(lam), np.array(spec_1), np.array(spec_2), np.array(spec_3),np.array(spec_4), np.array(spec_5), np.array(spec_6)
```

```python
spec_1 = np.interp(lam_1, lam_1, spec_1_1)
spec_2 = np.interp(lam_1, lam_1, spec_2)
spec_3 = np.interp(lam_1, lam_1, spec_3)
spec_4 = np.interp(lam_1, lam_1, spec_4)
spec_5 = np.interp(lam_1, lam_1, spec_5)
spec_6 = np.interp(lam_1, lam_1, spec_6)
lam = np.linspace(min(lam_1), max(lam_1), len(spec_2))
```

```python
fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(12,6))
axs.set_title('Yıldız Tayfı ve Evreleri', fontsize= 18)

axs.plot(lam , (spec_6+5), 'g', label ='Evre 0.9', linewidth= 0.5)
axs.plot(lam , (spec_5+4), 'c', label ='Evre 0.75', linewidth= 0.5)
axs.plot(lam , (spec_4+3), 'y', label ='Evre 0.6', linewidth= 0.5)
axs.plot(lam , (spec_3+2), 'k', label ='Evre 0.5', linewidth= 0.5)
axs.plot(lam , (spec_2+1), 'r', label ='Evre 0.25', linewidth= 0.5)
axs.plot(lam, spec_1, 'b', label ='Evre 0.1', linewidth= 0.5)

axs.set_xlabel('Dalga Boyu $[\AA]$', fontsize= 13)
axs.set_ylabel('Akı Yoğunluğu', fontsize= 13)
axs.grid(True)
axs.legend(loc="upper right")
#plt.savefig(fname='image_1.1.png', facecolor="white")
plt.show()
```
